# Step 3 : Data Preparation

```python
# Converting categorical variables having 2 values to 1 or 0
cars['fueltype'] = cars['fueltype'].map({'gas': 0, 'diesel': 1})
cars['aspiration'] = cars['aspiration'].map({'std': 0, 'turbo': 1})
cars['doornumber'] = cars['doornumber'].map({'two': 0, 'four': 1})
cars['enginelocation'] = cars['enginelocation'].map({'front': 0, 'rear': 1})
cars.head()
```

```python
def get_dummies(df, var):
    status = pd.get_dummies(df[var], drop_first = True)
    df = pd.concat([df, status], axis = 1)
    df.drop([var], axis = 1, inplace = True)
    return df
```

```python
cars = get_dummies(cars, 'carbody',)
cars.head()
```

```python
cars = get_dummies(cars, 'drivewheel')
cars = get_dummies(cars, 'enginetype')
cars = get_dummies(cars, 'cylindernumber')
cars = get_dummies(cars, 'fuelsystem')
cars.head()
```

```python
cars.columns
```

```python
cars['CompanyName'].unique()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x = 'CompanyName', y = 'price', data = cars)
plt.xticks(rotation=90)
plt.show()
```

```python
# Converting CompanyName to to high, medium and low as per price of the cars
cars['price'] = cars['price'].astype('int')
temp = cars.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')

bins = [0,10000,20000,40000]
company_bin=['Low','Medium','High']
cars['company_segment'] = pd.cut(cars['price'],bins,right=False,labels=company_bin)
cars.head()
```

```python
# Dropping company name and introductind dummy for company_segment
cars.drop('CompanyName', axis = 1, inplace = True)
```

```python
cars = get_dummies(cars, 'company_segment')
cars.head()
```

# Step 4: Splitting the Data into Training and Testing Sets

```python
# Splitting the dataset into train and test dqataset
np.random.seed(0)
cars_train, cars_test = train_test_split(cars, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
cars_train.shape
```

```python
cars_test.shape
```

#### Rescaling the Features

```python
scaler = MinMaxScaler()
```

```python
# applying scalar to all numerical variables
num_vars = ['wheelbase','carlength','carwidth','carheight', 'curbweight', 'enginesize', 'boreratio','stroke','compressionratio', 'horsepower','peakrpm','citympg', 'highwaympg','price']

cars_train[num_vars] = scaler.fit_transform(cars_train[num_vars])
# cars.columns
```

```python
cars_train.head()
```

```python
cars_train.describe()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (30, 20))
sns.heatmap(cars_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

#### Dividing into X and Y sets for the model building

```python
y_train = cars_train.pop('price')
X_train = cars_train
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 25)       
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

# Step 5: Building Linear model

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding constant variable
X_train_lm = sm.add_constant(X_train_rfe)
```

```python
# Fitting the model
lm = sm.OLS(y_train,X_train_lm).fit()
```
